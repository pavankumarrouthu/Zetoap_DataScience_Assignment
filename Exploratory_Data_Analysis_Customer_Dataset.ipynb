```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv('Customers.csv')
```

***Customer Distribution by region***

```python
data['SignupDate'] = pd.to_datetime(data['SignupDate'])
```

```python
plt.figure(figsize=(10,6))
data['Region'].value_counts().plot(kind = 'bar')
plt.title("CDR")
plt.xlabel("Region")
plt.ylabel("No of customers")
plt.show()
```

***Customer SignUp Trends***

```python
plt.figure(figsize=(10,6))
data['SignupDate'].value_counts().sort_index().plot(kind='line')
plt.title('CST Over time')
plt.xlabel('Date')
plt.ylabel('No of SighUps')
plt.show()

```

***Customer Retention Analysis***

```python
data['Time_span'] = (pd.Timestamp.now()-data['SignupDate']).dt.days
plt.figure(figsize=(10,6))
sns.histplot(data['Time_span'],bins=30,kde=True)
plt.title('Customer Tenrure Distribution ')
plt.xlabel("Tenure in Days")
plt.ylabel("No of Customers")
plt.show()
```
